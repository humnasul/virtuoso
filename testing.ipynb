```python
# First cell: Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Second cell: Creating a basic dataset
data = {'X': np.arange(1, 11), 'Y': np.random.randint(1, 100, 10)}
df = pd.DataFrame(data)

# Third cell: Displaying the dataset
print(df)

# Fourth cell: Plotting the dataset
plt.plot(df['X'], df['Y'], marker='o', linestyle='-')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Basic Line Plot')
plt.show()
```
